PREDICT CANCER AS MALLIGNENT OR BENIGN

```python
from sklearn.datasets import load_breast_cancer
Data = load_breast_cancer()
```

```python
print(Data.DESCR)
```

```python
features = Data.data
target = Data.target

print(features.shape)
print(target.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split ( features, target, test_size=0.2)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense
```

```python
cancer_model = Sequential()
cancer_model.add(Dense(32, input_dim = 30, activation = 'relu'))
cancer_model.add(Dense(64, activation = 'relu')) # hidden layer
cancer_model.add(Dense(1, activation = 'sigmoid'))
```

```python
cancer_model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cancer_model.summary()
```

```python
cancer_model.fit(X_train, Y_train, epochs=10)
```

```python
scores = cancer_model.evaluate(X_test,Y_test)
print(scores) # returns loss and accuracy
```

```python
predictions = cancer_model.predict(X_test)
label =[]
for pred in predictions:
  if pred>=0.5:
    print("Mallignent")
  else:
    print("Benign")
```
